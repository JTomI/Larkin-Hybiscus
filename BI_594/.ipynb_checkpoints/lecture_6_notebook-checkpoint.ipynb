This code produces a graph based on data from Forchhammer and Lindahl (1971) that measures protein and RNA levels in E. coli at different per capita growth rates.

```python
# import the relevant modules

import numpy as np
import math
import matplotlib as mpl
import scipy as sp
import scipy.special
import matplotlib.pyplot as plt
import csv as csv
import pandas as pd
```

```python
# this is the url containing the data
Forchhammer_Lindahl_url = 'http://www.larkinlab.com/s/Forchhammer_Lindahl_1971_Table_1.csv'

# this imports the data into a pandas dataframe
df = pd.read_csv(Forchhammer_Lindahl_url)
# take a look at the data
print(df)
```

```python
# many microbiologists have noticed that RNA/protein ratio increases with per capita growth rate.
# let's make a new column with that ratio

df['tRNA_Ratio'] = df['tRNA_ug_per_mg']/df['protein_ug_per_mg']
df['rRNA_Ratio'] = df['rRNA_ug_per_mg']/df['protein_ug_per_mg']
print(df)
```

```python
# here we'll make a graph
fig,ax = plt.subplots(figsize=(15,10))
# ax.scatter(df['protein_ug_per_mg'],df['tRNA_ug_per_mg'])
# ax.scatter(df['protein_ug_per_mg'],df['rRNA_ug_per_mg'])
ax.plot(df['protein_ug_per_mg'],df['rRNA_Ratio']) 
ax.plot(df['protein_ug_per_mg'],df['tRNA_Ratio']) 
```
