**Solving differnetial equations  in python.** This text looks long and complicated, but I don't think it should be too bad. The goal is just to see how we can use python to numerically solve an equation and plot the solution. During the course, we will often have to solve ordinary differential equations (ODEs). These are equations that relate a function, like the density of cells in a bacterial culture over time, to its time derivatives (i.e. its rates of change). This is useful because it's often simple to write down an equation as a model, then we can solve it for different parameter values and compare the solutions to experimental data. 

To repeat, a differential equation contains a function and its derivatives, and the goal is to solve for the function to understand how the system evolves in time. To properly define and solve a differential equation, we need the equation, and the _initial value of the function_. For example, for an equation modeling bacterial growth, we will need the equation that describes the growth and the number of bacteria at the beginning of the experiment (i.e. at _t = 0_).

Here is an example where y represents the number of cells in a bacterial culture and t is time in hours.<br />
<center>\begin{equation}\frac{dy}{dt} = 1.5y\end{equation} \begin{equation}y(0) = 1000\end{equation}</center><br />

This means at time = 0, there are 1 thousand cells in the tube. Then as the cells grow, the rate of population growth (i.e. $\frac{dy}{dt}$) is directly proportional to the population size with a proportionality factor of 1.5. As we've said in lecture, a few cells don't produce much growth; a lot of cells produce a lot of growth. We would like to solve for the number of cells over time. This is called an _initial value problem_. Python has a nice way to solve it automatically! Let's check it out.

This method uses a package from the scipy python package called scipy.integrate. Specifically, we will use a method within scipy.integrate: scipy.integrate.solve_ivp (for "solve initial value problem"). Here are the steps to solving an initial value problem with solve_ivp:

1. Write out the ODE by hand in the form<br /><center>\begin{equation}\frac{dy}{dt} = f(t,y)\end{equation} \begin{equation}y(0) = y_{0}\end{equation}</center><br />
2. Import solve_ivp from the scipy.integrate module (e.g. "from scipy.integrate import solve_ivp").
3. Define _f_ on the right side of the equation.
4. Create an array of the initial and final time points over which you'd like to solve the equation. This must be a numpy array.
5. Create an array of the initial condition $y(0) = y_{0}$.
5. Solve the equation with solve_ivp.

Let's solve the above equation. Below I've gone through the solution. Read along with the code and comments. You can just run the code and then you'll set up a similar one. The technique we're using here may seem overkill for a simple equation like the one above, but we'll see later in the course that you can quickly solve
complicated equations to get beautiful solutions.

```python
# 2. import solve_ivp
from scipy.integrate import solve_ivp

# here's a constant representing the per-capita growth rate, mu
per_capita_growth = 1.5

# 3. define the function
def f_1(t,y):
    return per_capita_growth*y

# 4. define the time interval over which to solve and the initial condition. Here we're doing 0 to 5 hours.
time_interval = np.array([0,5]) # this array contains the first time point and the last time point.
solution_times = np.linspace(time_interval[0], time_interval[1], 100) # this array contains 100 evenly spaces time points over which we'll solve the equation.

# 5. initial condition
y_0 = np.array([1000.0]) # this is the initial condition. Even though it's a single number, we need to define it as a numpy array containing one number.

# 6. solve!
solution_y = solve_ivp(f_1, time_interval, y_0, t_eval = solution_times) # use solve_ivp() to solve the equation. this variable will contain the solution.

# note the order in which we have to enter the arguments for solve_ivp().

# here is the time axis and y-axis for the solution from solve_ivp().
time_t = solution_y.t # for solve_ivp, we always get the time axis from the solution with .t and the the y-axis from .y as below, no matter what we name the solution variable.
y_t = solution_y.y[0] # for a single differential equation, the solution is the first element of of .y, i.e. solution_y.y[0]. We will see later that we can use this method to solve systems of more than one equation.

# now plot the solution
plt.figure()
plt.plot(time_t, y_t, '-b')

plt.xlabel('time (hours)')
plt.ylabel('number of cells')
```

What kind of growth does this appear to be from the solution to the equation? Hint: we're talking about this kind of growth a lot, haha. From the equation, you'll see that this kind of growth arises when the rate of population growth is proportional to the population size.

Now you'll solve a similar equation. In a bacterial growth curve, there is an exponential phase where the cells reproduce with a constant doubling time, then a stationary phase where their growth slows down and stops. After the stationary phase, we generally see a _death phase_, where the number of cells declines. During this period, the rate of decline of the number of cells is proportional to the number of cells. The logic is similar to the first equation: with a large number of cells dying, the rate of decline will be fast. With few cells, it will be slow. We can represent this with an equation, this time using _N_ to represent the number of cells:<br />
<center>\begin{equation}\frac{dN}{dt} = -N\end{equation}</center><br />

If at the beginning of the death phase there are 1 billion cells in the culture, fill in the following code to solve the equation and make a graph for the number of cells in this tube as a function of time.

```python
# uncomment the code below and fill in more to solve this problem.

# define the function describing the population's rate of change (right side of the above equation)
def f_death(t,N):
    return -N

# define the time interval over which to solve and the initial condition.
interval = np.array([0,5])
death_solution_times = np.linspace(interval[0],interval[1],100)

# define the initial value, i.e. the number of cells in the culture at the beginning of the death phase.
N_0 = np.array([1e9])

# solve with solve_ivp()
solution_N = solve_ivp(f_death,interval,N_0,t_eval=death_solution_times)

# print(solution_N)
# get the t-axis and N values (i.e. the y-axis) from the above variable with .t and .y[0].
time_t = solution_N.t
N_t = solution_N.y[0]

# now plot the solution with labeled axes.
fig,ax = plt.subplots(figsize=(15,10))
ax.set_title('Population of Living Cells During Death Phase')
ax.scatter(time_t, N_t)
ax.set_xlabel('time (hours)')
ax.set_ylabel('number of cells')
```
