```python
import numpy as np
import math
import matplotlib as mpl
import scipy as sp
import scipy.special
import matplotlib.pyplot as plt
import pandas as pd
# import the solve_ivp method to solve the equation
from scipy.integrate import solve_ivp as solve
```

Here we'll solve for the gene expression dynamics of a gene that is unregulated, but then becomes perfectly repressed after 8 hours when the cell detects a signal that activates a repressor. The cell is growing with a doubling time of 1 hour.

```python
# first define key parameters
max_transc_rate = 3600 # from transcription rate with molecules/cell and time in seconds
mRNA_deg_k = math.log(2)/0.1 # time is in hours! mRNA half-life is ~6 minutes, which is 0.1 hours.
max_transl_rate = 5142 # maximim possible translation rate for mRNA
prot_deg_k = math.log(2)/1.0 # protein degration rate is due to dilution; this is ln(2) over the doubling time.
switch_time = 8 # this is the time at which the repressor gets turned on
```

```python
# define the gene expression equations that we'll solve
def gene_expr_func(t,g):
    if (t<switch_time):
        dm_dt = max_transc_rate - mRNA_deg_k*g[0]
        dp_dt = max_transl_rate*g[0]-prot_deg_k*g[1]
    else:
        dm_dt = - mRNA_deg_k*g[0]
        dp_dt = max_transl_rate*g[0]-prot_deg_k*g[1]
    return dm_dt,dp_dt
```

```python
# define the time interval over which to solve
time_interval = np.array([0,12])
solution_times = np.linspace(time_interval[0],time_interval[1],1000)
# print(solution_times)

# set up the initial condition
g_0 = np.array([0,0])

# solve the equation
solution_g = solve(gene_expr_func, time_interval,g_0,t_eval=solution_times)
# extract the solution arrays
t = solution_g.t
mRNA = solution_g.y[0]
prot = solution_g.y[1]
```

```python
# plot the solutions as graphs

fig = plt.figure(figsize= (6,12))

m_ax = fig.add_subplot(2,1,1)
m_ax.plot(t, mRNA,'red')
m_ax.set_title('mRNA Growth')
m_ax.set_ylabel('Concentration')
m_ax.set_xlabel('Time')

p_ax = fig.add_subplot(2,1,2)
p_ax.plot(t, prot,'green')
p_ax.set_title('Protein Growth')
p_ax.set_ylabel('Concentration')
p_ax.set_xlabel('Time');
```
