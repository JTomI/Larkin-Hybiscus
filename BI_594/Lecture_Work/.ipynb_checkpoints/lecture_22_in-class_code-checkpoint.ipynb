```python
# import the appropriate modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy as sp
import math

figsize=(7.5,5)
fontsize=20;
```

```python
# import the data from here: 'https://octagon-seal-xej4.squarespace.com/s/Heukelekian_Heller_1940_growth_data.csv'
df = pd.read_csv('https://octagon-seal-xej4.squarespace.com/s/Heukelekian_Heller_1940_growth_data.csv')
print(df)
# make the graph below
fig, ax = plt.subplots(figsize=figsize)
ax.loglog(df["food_conc_mg_kg"],df["no_bead_density"],label="No Bead");
ax.loglog(df["food_conc_mg_kg"],df["bead_density"],label="Bead");
ax.set_xlabel("Nutrient Concentration (mg/kg)",fontsize=fontsize)
ax.set_ylabel("Net Growth (CFUs)",fontsize=fontsize)
ax.set_title('Surface Dependance of Bacterial Growth',fontsize=fontsize)
ax.legend(fontsize=fontsize);
```
